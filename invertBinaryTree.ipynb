```python
# input
# tree =       1
#            /   \
#          2      3
#         /  \   /  \
#        4    5 6    7
#      /   \
#     8     9

# output
# tree =      1
#           /   \
#          3     2
#        /  \   /  \
#       5    4 7    6
#                 /   \
#                9     8
# credit: https://www.youtube.com/watch?v=vHKzIPwWQkg
# https://www.tutorialspoint.com/python_data_structure/python_binary_tree.htm
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    def printTree(self):
        if self.left:
            self.left.printTree()
        print(self.data)
        if self.right:
            self.right.printTree()
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
    def invertBinaryTree(self, tree):
        if tree:
            left = tree.left
            right = tree.right
            tree.left = right
            tree.right = left
            self.invertBinaryTree(tree.left)
            self.invertBinaryTree(tree.right)
root = Node(9)
root.insert(3)
root.insert(2)
root.insert(5)
root.insert(4)
root.insert(6)
root.insert(8)
root.insert(1)
root.insert(7)
root.printTree()
root.invertBinaryTree(root)
root.printTree()
```
